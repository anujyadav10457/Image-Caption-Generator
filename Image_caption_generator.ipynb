```python

import os
import string
import numpy as np
import pandas as pd
from tqdm import tqdm
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical, plot_model
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding, Dropout, Add
from tensorflow.keras.models import Model
from tensorflow.keras.utils import plot_model
from keras_preprocessing.image import load_img, img_to_array
import nltk
nltk.download('punkt')

# Define the paths
images_path = "flickr8k/Images"
captions_path = "flickr8k/captions.txt"
```

```python
# Load captions
def load_captions(captions_path):
    with open(captions_path, 'r') as f:
        captions = f.read()
    return captions

# Create a dictionary of image to captions
def preprocess_captions(captions):
    mapping = {}
    for line in captions.split('\n'):
        tokens = line.split()
        if len(tokens) < 2:
            continue
        image_id, image_caption = tokens[0], tokens[1:]
        image_id = os.path.splitext(image_id)[0]  # Strip file extension
        image_caption = ' '.join(image_caption)
        if image_id not in mapping:
            mapping[image_id] = []
        mapping[image_id].append(image_caption)
    return mapping

captions = load_captions(captions_path)
captions = preprocess_captions(captions)

# Clean captions
def clean_captions(captions):
    table = str.maketrans('', '', string.punctuation)
    for key, desc_list in captions.items():
        for i in range(len(desc_list)):
            desc = desc_list[i]
            desc = desc.split()
            desc = [word.lower() for word in desc]
            desc = [word.translate(table) for word in desc]
            desc = [word for word in desc if len(word) > 1]
            desc = [word for word in desc if word.isalpha()]
            desc_list[i] = ' '.join(desc)

# Convert captions to sequences of word indices
def create_tokenizer(captions):
    desc_list = []
    for key in captions.keys():
        [desc_list.append(d) for d in captions[key]]
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(desc_list)
    return tokenizer

def max_length(captions):
    desc_list = []
    for key in captions.keys():
        [desc_list.append(d) for d in captions[key]]
    return max(len(d.split()) for d in desc_list)

captions = load_captions(captions_path)
captions = preprocess_captions(captions)
clean_captions(captions)
tokenizer = create_tokenizer(captions)
vocab_size = len(tokenizer.word_index) + 1
max_length = max_length(captions)

```

```python
# Extract features using VGG16
def extract_features(directory):
    model = VGG16()
    model = Model(inputs=model.inputs, outputs=model.layers[-2].output)
    features = {}
    for name in tqdm(os.listdir(directory)):
        filename = os.path.join(directory, name)
        image = load_img(filename, target_size=(224, 224))
        image = img_to_array(image)
        image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
        image = preprocess_input(image)
        feature = model.predict(image, verbose=0)
        image_id = name.split('.')[0]
        features[image_id] = feature
    return features
```

```python
features = extract_features(images_path)
```

```python
import tensorflow as tf

def preprocess_caption(tokenizer, max_length, caption, vocab_size):
    seq = tokenizer.texts_to_sequences([caption])[0]
    X2, y = [], []
    for i in range(1, len(seq)):
        in_seq, out_seq = seq[:i], seq[i]
        in_seq = tf.keras.preprocessing.sequence.pad_sequences([in_seq], maxlen=max_length)[0]
        out_seq = tf.keras.utils.to_categorical([out_seq], num_classes=vocab_size)[0]
        X2.append(in_seq)
        y.append(out_seq)
    return X2, y

def data_generator(captions, images, tokenizer, max_length, vocab_size):
    while True:
        for key, image in images.items():
            if key not in captions:
                continue
            for caption in captions[key]:
                X2, y = preprocess_caption(tokenizer, max_length, caption, vocab_size)
                for in_seq, out_seq in zip(X2, y):
                    yield (image, in_seq), out_seq

```

```python
def create_tf_dataset(captions, images, tokenizer, vocab_size, max_length, batch_size):
    dataset = tf.data.Dataset.from_generator(
        lambda: data_generator(captions, images, tokenizer, max_length, vocab_size),
        output_signature=(
            (
                tf.TensorSpec(shape=(4096,), dtype=tf.float32),  # Image features
                tf.TensorSpec(shape=(max_length,), dtype=tf.int32)  # Input sequences
            ),
            tf.TensorSpec(shape=(vocab_size,), dtype=tf.float32)  # Output sequences
        )
    )
    
    dataset = dataset.shuffle(buffer_size=10000)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(buffer_size=tf.data.AUTOTUNE)
    
    return dataset

# Initialize the tokenizer
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts([caption for key in captions for caption in captions[key]])

# Create the dataset
batch_size = 64
train_dataset = create_tf_dataset(captions, features, tokenizer, vocab_size, max_length, batch_size)


```

```python
from tensorflow.keras.layers import Input, Dense, Dropout, Embedding, LSTM, Add
from tensorflow.keras.models import Model

def define_model(vocab_size, max_length):
    # Feature extractor model
    inputs1 = Input(shape=(4096,))
    fe1 = Dropout(0.5)(inputs1)
    fe2 = Dense(256, activation='relu')(fe1)
    
    # Sequence model
    inputs2 = Input(shape=(max_length,))
    se1 = Embedding(vocab_size, 256, mask_zero=True)(inputs2)
    se2 = Dropout(0.5)(se1)
    se3 = LSTM(256)(se2)
    
    # Decoder model
    decoder1 = Add()([fe2, se3])
    decoder2 = Dense(256, activation='relu')(decoder1)
    outputs = Dense(vocab_size, activation='softmax')(decoder2)
    
    # Merge the models
    model = Model(inputs=[inputs1, inputs2], outputs=outputs)
    model.compile(loss='categorical_crossentropy', optimizer='adam')
    
    return model

model = define_model(vocab_size, max_length)
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import Callback

class PrintEpochCallback(Callback):
    def on_epoch_end(self, epoch, logs=None):
        loss = logs.get('loss')
        print(f"Epoch {epoch + 1}: loss = {loss:.4f}")

print_epoch_callback = PrintEpochCallback()

# Train the model
model.fit(train_dataset, epochs=20, verbose=2, callbacks=[print_epoch_callback])
```

```python
def generate_caption(model, tokenizer, photo, max_length):
    in_text = 'startseq'
    for _ in range(max_length):
        sequence = tokenizer.texts_to_sequences([in_text])[0]
        sequence = pad_sequences([sequence], maxlen=max_length)
        yhat = model.predict([photo,sequence], verbose=0)
        yhat = np.argmax(yhat)
        word = tokenizer.index_word.get(yhat)
        if word is None:
            break
        in_text += ' ' + word
        if word == 'endseq':
            break
    return in_text

# Load and preprocess the image
def load_image(filename):
    image = load_img(filename, target_size=(224, 224))
    image = img_to_array(image)
    image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
    image = preprocess_input(image)
    return image

filename = 'path/to/new_image.jpg'
photo = load_image(filename)
feature = model.predict(photo, verbose=0)
caption = generate_caption(model, tokenizer, feature, max_length)
print(caption)

```
